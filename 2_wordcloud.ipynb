```python
import os
from konlpy.tag import Okt
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm 
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

```python
# 한 글자 이상 명사만 저장
words = []
with open('인공지능youtube댓글.csv', 'r', encoding='utf-8') as f:
    text = f.read()
okt = Okt()
nouns = okt.nouns(text) 
for i in nouns:
    if i in ['말','힘','삶']: # 세개의 고빈도 한글자 단어 제외
        words.append(i)
    elif len(i) > 1:
        words.append(i)
words
```

```python
# 단어 개수 세기

text_cnt = Counter(words)
sorted_dict = sorted(text_cnt.items(), key = lambda item: item[1], reverse = True)
print(sorted_dict)
```

```python
# 단어 빈도표 그리기

import matplotlib.pyplot as plt
import nltk
import warnings
warnings.filterwarnings("ignore")
import matplotlib as mpl
from matplotlib import font_manager

%config InlineBackend.figure_format = 'retina'
fontpath = 'content/sample_data/NanumGothic.ttf'
plt.rc('font', family='NanumGothic') 
# mpl.font_manager._rebuild()



#plt.figure(figsize=(12,6))
plt.figure(figsize=(20,10))
words = nltk.Text(words, name='단어 빈도수')
words.plot(50) # 50개만
plt.show()
```

```python
# 단어 빈도표를 보고 일부 단어 불용어 처리
stopword = ['진짜','영상', '정말', '오늘', '지금', '때문', '그냥','보고']
text = [i for i in words if i not in stopword]
text_cnt = Counter(text)
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python

wordcloud = WordCloud(font_path = fontpath, 
                      background_color='white',
                      # colormap = "Accent_r", 
                      # mask = img_mask, 
                      random_state = 20, 
                      max_words = 200).generate_from_frequencies(text_cnt)

plt.figure(figsize = (10, 10))
plt.imshow(wordcloud, interpolation = "bilinear")        
plt.axis("off")

plt.savefig("result_AI.png")
plt.show()
```

```python
#워드클라우드로 나타내기, 자주 언급된 키워드 시각화

import nltk 
from konlpy.tag import Kkma
from konlpy.tag import Twitter

import matplotlib.pyplot as plt
from wordcloud import WordCloud
from wordcloud import STOPWORDS 
import numpy as np
from PIL import Image
from wordcloud import ImageColorGenerator

stopwords = set(STOPWORDS)    
img_mask = np.array(Image.open("로봇.jpg")) 
image_colors = ImageColorGenerator(img_mask)
stopwords = set(STOPWORDS)    
FONTPATH =  "NanumGothic.ttf"

wordcloud = WordCloud(max_font_size = 100, max_words = 1000, background_color='white', stopwords=stopwords, random_state=10,mask=img_mask, color_func=image_colors, font_path = FONTPATH)
wc = wordcloud.generate_from_frequencies(text_cnt)

plt.figure(figsize = (20,20))
plt.imshow(wc, interpolation = "lanczos")
plt.axis("off")
plt.show()
```
