```python
import pandas as pd
import numpy as np
import gensim
import gensim.corpora as corpora

from konlpy.tag import Komoran
tagger = Komoran()
```

## POS Tagging

```python
talmud_text = list(df)
```

```python
def tokenize(data):
    result = []
    for sent in data:
        tokens = komoran.pos(sent)
        new_tokens = []
        for token in tokens:
            if token[1] in ['NNG', 'NNP']:
                new_token = token[0]
                new_tokens.append(new_token)
            elif token[1] in ['VV', 'VA']:
                new_token = token[0]
                new_token += "다"
                new_tokens.append(new_token)
            else:
                continue
        result.append(new_tokens)
    return(result)
```

```python
talmud_tokens = tokenize(talmud_text)
```

## Topic Modeling(LDA)

```python
def sent_to_words(sentences):
    for sentence in sentences:
        yield(gensim.utils.simple_preprocess(str(sentence), deacc=True))
```

```python
lda_tokens = list(sent_to_words(talmud_tokens))
```

```python
id2word = corpora.Dictionary(lda_tokens)
```

```python
tdf_mat = [id2word.doc2bow(text) for text in lda_tokens]
```

```python
lda_model = gensim.models.ldamodel.LdaModel(corpus = tdf_mat,
                                           id2word = id2word,
                                           num_topics = 10, 
                                           random_state = 0,
                                           alpha='auto',
                                           per_word_topics = True)
```

```python
lda_model.show_topics(10, 20)
```

```python
def get_doc_topic(corpus, model):
    doc_topic = list()

    for doc in corpus:
        doc_topic.append(model.get_document_topics(doc))
       
    doc_topic = [dict(i) for i in doc_topic]
    doc_topic = pd.DataFrame(doc_topic)
    doc_topic.fillna(value = 0, inplace = True)
    return doc_topic
```

```python
dt_mat = get_doc_topic(tdf_mat, lda_model)
```

## Co-Occurence

```python
from collections import defaultdict
```

```python
def co_occurrence(sentences, window_size):
    d = defaultdict(int)
    vocab = set()
    for text in sentences:
        for i in range(len(text)):
            token = text[i]
            vocab.add(token)  # add to vocab
            next_token = text[i+1 : i+1+window_size]
            for t in next_token:
                key = tuple(sorted([t, token]))
                d[key] += 1

    # formulate the dictionary into dataframe
    vocab = sorted(vocab) # sort vocab
    df = pd.DataFrame(data=np.zeros((len(vocab), len(vocab)), dtype=np.int16),
                      index=vocab,
                      columns=vocab)
    for key, value in d.items():
        df.at[key[0], key[1]] = value
        df.at[key[1], key[0]] = value
    return df
```

```python
co_occ_df = co_occurrence(talmud_tokens, 4) # window 지정
```

```python
co_occ_df = co_occ_df.reindex(w2v_words)
```

```python
co_occ_df = co_occ_df[w2v_words]
```

```python
co_occ_df
```

```python
nlargest = 10
co_occ_order = np.argsort(-co_occ_df.values, axis=1)[:, :nlargest]
co_occ_words = pd.DataFrame(co_occ_df.columns[co_occ_order], 
                      columns=['top{}'.format(i) for i in range(1, nlargest+1)],
                      index=co_occ_df.index)
```

## Clustering

```python
from sklearn .cluster import KMeans
from sklearn.metrics import silhouette_score, silhouette_samples
```

```python
km_wv = w2v_model[w2v_model.wv.vocab]
```

```python
len(km_wv)
```

```python
k_range = [2, 3, 4, 5, 6, 7, 8, 9, 10]

for k in k_range:
    clusterer = KMeans(n_clusters = k)
    preds = clusterer.fit_predict(km_wv)
    centers = clusterer.cluster_centers_

    score = silhouette_score(km_wv, preds, metric = 'euclidean')
    print("For n_clusters = {}, silhouette score is {})".format(k, score))
```

```python
kmc = KMeans(n_clusters = 4)
```

```python
kmc_preds = kmc.fit_predict(km_wv)
```

```python
result_kmc = list(kmc_preds)
```
